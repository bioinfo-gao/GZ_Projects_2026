因为 Positron 是基于 VS Code (Code-OSS) 核心构建的，所以它查找和配置快捷键的逻辑与 VS Code 完全相同。

以下是查找并设置 .ipynb 中移动 cell 快捷键的具体步骤：

1. 打开快捷键设置面板
最快方式： 依次按下组合键 Ctrl + K，松开后紧接着按 Ctrl + S。

通过命令面板： 按下 Ctrl + Shift + P 打开命令面板，输入 Keyboard Shortcuts（若界面已汉化则输入“键盘快捷方式”），并点击打开。

2. 搜索目标命令
在快捷键面板顶部的搜索框中，直接输入以下关键词过滤：
notebook move cell

此时列表中会出现你需要的核心命令：

Notebook: Move Cell Up (笔记本: 向上移动单元格)

Notebook: Move Cell Down (笔记本: 向下移动单元格)

3. 查看或自定义快捷键
默认状态： 在 Positron 中，移动 cell 的默认快捷键通常是 Alt + ↑（向上方向键）和 Alt + ↓（向下方向键）。

如何修改： 找到命令后，双击该条目，或者点击条目左侧出现的铅笔图标 ✏️。接着在弹出的输入框中，直接按下你想绑定的新组合键（比如 Ctrl + Shift + Up 等），最后按 Enter 确认保存。
