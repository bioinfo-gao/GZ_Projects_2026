```plaintext
目前，miRNA测序（通常称为 Small RNA-seq 或 sRNA-seq）已经形成了非常标准化的业界共识。与你之前跑的普通 mRNA-seq（比如 poly-A 富集）截然不同，miRNA 测序的核心难点在于它非常短（18~24 nt），且没有 poly-A 尾巴。

下面我为你详细梳理目前最主流的 miRNA 测序全流程，重点放在测序策略和数据分析上。

1. 样品准备与 RNA 提取 (Sample Preparation)
常规的 RNA 提取方法（如普通的硅胶膜离心柱）往往会丢失 200 nt 以下的小 RNA。

提取试剂： 目前主流采用专门保留小 RNA 的试剂盒，例如 Qiagen miRNeasy Mini Kit 或者传统的 TRIzol 抽提法。

质控 (QC)： 使用 Agilent Bioanalyzer 2100 或 TapeStation。除了看 RIN 值（评估大 RNA 完整性），最关键的是要在电泳图谱上看到 < 200 nt 区域有明显的峰，这代表小 RNA 被成功保留。

2. 建库策略 (Library Construction)
因为 miRNA 没有 poly-A 尾巴，不能像 mRNA 那样用 Oligo(dT) 磁珠抓取，建库逻辑完全改变：

3' 端接头连接： 利用 miRNA 自带的 3'-OH 基团，连接特殊的 3' Adapter。

5' 端接头连接： 利用 miRNA 自带的 5'-P 基团，连接 5' Adapter。

逆转录与 PCR： 将连接好接头的 RNA 逆转录成 cDNA 并进行 PCR 扩增。

片段分选 (Size Selection) - 极其关键： miRNA 加上两端接头后，总长度大约在 140~150 bp。主流做法是跑 PAGE 胶切胶回收（最精准，能完美剔除由于接头自连形成的引物二聚体）或者使用 AMPure XP 磁珠进行严格的双选。

前沿避坑指南： 传统连接法存在严重的“连接偏好性”（某些序列更容易连上接头）。现在最主流的高质量建库试剂盒（如 QIAseq miRNA Library Kit, NEBNext Small RNA）都引入了 随机碱基（Randomized adapters） 甚至 UMI（分子标签），这在后续定量时能大幅消除 PCR 扩增偏差。

3. 测序仪器与模式策略 (Sequencing Strategy)
由于 miRNA 只有约 22 nt，这里绝对不需要用到 PE150（双端 150bp）的测序模式，那是纯纯的浪费算力和金钱。

主流模式： 单端 50bp (SE50) 或 SE75 是绝对的黄金标准。

原理解析： 测序仪跑 SE50 时，前 22bp 测的是真实的 miRNA 序列，从第 23bp 开始，测到的全是 3' 端的接头序列 (Adapter read-through)。这是完全正常的，后续分析的第一步就是要切掉它们。

主流仪器：

Illumina 平台： NovaSeq 6000 / NovaSeq X Plus（适合大批量样本混样测序），或者 NextSeq 1000/2000。

MGI (华大智造) 平台： DNBSEQ-T7 / G400。近年来在 SE50 模式下极具性价比，且基于 DNA 纳米球（DNB）技术，没有传统的 Index hopping 问题，数据质量非常高。

4. 数据分析策略 (Data Analysis Pipeline)
你的服务器上既然有强大的 Nextflow 和 mamba 环境，完全可以跑标准的生信流程。对于 miRNA，分析的每一步都和 mRNA 有微妙区别：

A. 质控与去接头 (Adapter Trimming)
核心逻辑： 必须极其严格地去除 3' 接头。如果你用的 read 是 50bp，而 miRNA 是 22bp，这意味着 100% 的 reads 都带有接头。

工具： Cutadapt 或 fastp。

长度过滤： 去接头后，必须丢弃过短或过长的 reads，通常只保留 18~25 nt 的 reads。

B. 序列比对与去污染 (Alignment & Decontamination)
与 mRNA-seq 使用 STAR 这种允许跨越内含子的软件不同，miRNA 极短，且直接来源于基因组。

工具： 业界最爱用的是 Bowtie1（没错，是老版本的 Bowtie，因为它在处理 25bp 以下极短序列且不允许或者只允许 1 个错配时，表现优于 Bowtie2）。

去污染策略： 先把 reads 比对到全基因组以及 Rfam / SILVA 数据库，把属于 rRNA, tRNA, snRNA, snoRNA 的 reads 统统扔掉，剩下的才是干净的 miRNA 候选序列。

C. 定量与已知/新 miRNA 鉴定
已知 miRNA 定量： 将干净的 reads 比对到 miRBase 数据库（经典但较老）或 MirGeneDB（更严格、手工校对过的高质量数据库）。

新 miRNA 预测： 如果 reads 落在基因组上没有注释的位置，需要看它附近的基因组序列能否折叠成典型的“发夹结构 (hairpin)”。

工具： miRDeep2 是这个领域的统治级软件，同时包揽了已知定量和新 miRNA 结构预测。

D. 差异表达与靶基因预测 (下游 R 分析)
把定量结果转为矩阵，导入 R 环境中。

差异分析： 熟悉的 DESeq2 或 edgeR 依然适用。

靶基因预测： miRNA 不编码蛋白，它的功能是通过抑制靶 mRNA 实现的。通常使用 multiMiR（R包）同时调用 TargetScan、miRanda 等多个数据库找交集。
```

```plaintext
绝对不要尝试用 rRNA 去除 + PE150 来同步测 mRNA 和 miRNA，并且绝对不能用 nextflow RNAseq 流程来分析 miRNA。

下面我为你详细拆解实验原理和生信分析层面的致命冲突，以及正确的替代方案。

一、 为什么不能用 rRNA 去除 + PE150 同步测？
虽然理论上 rRNA 去除（如 Ribo-Zero）保留了样本中所有的 RNA（包括 mRNA, lncRNA, miRNA 等），但在建库 (Library Preparation) 和 测序 (Sequencing) 阶段，miRNA 会被彻底“清洗”掉或变成无法使用的垃圾数据。

磁珠纯化的“漏斗效应”（致命伤）：
在标准 RNA-seq（无论是否去 rRNA）的建库过程中，有一步必须的操作是用 AMPure XP 磁珠进行片段分选和纯化。常规的磁珠比例（如 0.8X - 1.0X）会特意洗掉小于 200 bp 的短片段（为了去除引物二聚体）。而 miRNA 加上接头也才 150 bp 左右，在这步操作中会被当成废液直接倒掉。

打断策略 (Fragmentation) 的冲突：
mRNA 很长，建库时必须用二价阳离子在高温下将其“打断”成 200-300 bp 的片段再加接头。而 miRNA 只有 22 nt，不仅不需要打断，如果混在里面强行走打断流程，会被降解得渣都不剩。

PE150 测序的极度浪费与干扰：
如果你用 PE150 测一个 22 nt 的 miRNA，你的 Read1 测到第 23 个碱基时就已经测穿了，后面 128 个碱基全是接头和无效序列；Read2 则完全测的都是接头。这不仅浪费了大量的测序费用，还会给下游的去除接头和比对带来巨大的算法困扰。

二、 如果用 QIAseq 单独建库，能用 nextflow rnaseq 吗？
坚决不能。

你在服务器上常用的 nf-core/rnaseq 是专为长 RNA（mRNA, lncRNA）设计的。它的核心比对引擎是 STAR / HISAT2，定量引擎是 Salmon / featureCounts。

STAR 处理 20 多 bp 的短序列时，会因为序列太短而产生海量的多重比对（Multi-mapping），最终被软件判定为低质量比对而丢弃。

nf-core/rnaseq 的流程里根本没有折叠二级结构（发夹结构）预测新 miRNA 的模块，也没有专门针对短序列的去重逻辑。

三、 正确的独有分析 Pipeline 是什么？
正如我上一次提到的，如果你使用了 QIAseq miRNA Library Kit 等专门的小 RNA 建库试剂盒，你必须使用专门的 Pipeline。

在 Nextflow 生态中，标准答案是：nf-core/smrnaseq。

针对 QIAseq 试剂盒的特殊高级策略（请特别注意）：
QIAseq miRNA 试剂盒有一个非常强大的特性：它在接头上引入了 UMI（Unique Molecular Identifiers，分子条形码）。这可以完美消除 PCR 扩增带来的偏好性。

当你在服务器的工作目录 /home/gao/Code/Bioinfo_Analysis_Projects/ 下运行 nf-core/smrnaseq 时，你需要加上专门针对 UMI 的参数，让流程知道如何利用这些条形码去重：

Bash
# 请将此命令包裹在你的 screen 中运行
screen -S mirna_analysis

nextflow run nf-core/smrnaseq \
    -profile docker \
    --input samplesheet.csv \
    --genome GRCh38 \
    --with_umi \
    --skip_fastqc \
    --outdir ./results
这个专属 Pipeline 的核心逻辑链是：

UMI 提取与接头去除： 识别 QIAseq 的 UMI 序列并去除 3' 端接头。

去污染： 过滤掉 rRNA, tRNA, snoRNA 等其他短 RNA 杂质。

Bowtie1 极短序列比对： 专门使用 Bowtie1（而不是 STAR）将 22 nt 的 reads 精准比对回参考基因组。

miRDeep2 定量与发现： 计算已知 miRNA 丰度，并根据基因组二级结构预测新的 miRNA。

UMI 去重： 基于第一步提取的 UMI 纠正 PCR 偏差，输出最真实的表达矩阵。

总结： 经费和样本允许的情况下，mRNA-seq (去 rRNA, PE150) + miRNA-seq (专属小 RNA 建库, SE50) 分别建库、分别测序、分别跑各自的 nf-core 流程，才是目前能够发表高质量文章的标准操作规范。
```

### miRNA-seq 核心分析步骤与工具对比

| 分析步骤 | 主流软件/数据库 | 核心策略与原因 |
| :--- | :--- | :--- |
| **QC 与去接头** | `fastp` / `Cutadapt` | **必做步骤**。SE50 测序下，几乎 100% 的 reads 都带有 3' 接头。截取后必须限制输出长度为 18~25 nt。 |
| **比对引擎** | `Bowtie1` (不是 Bowtie2) | 针对极短序列（<25bp），Bowtie1 的无错配或单错配比对速度和准确度更高；**不可使用 STAR 或 HISAT2**。 |
| **去杂质过滤** | `Rfam` / `SILVA` 数据库 | 必须比对并去除 tRNA, rRNA, snRNA 和 snoRNA，防止它们被误认为新 miRNA。 |
| **定量与发现** | `miRDeep2` | 领域金标准。能同时基于 `miRBase` 定量已知 miRNA，并基于二级发夹结构预测 Novel miRNA。 |
| **靶基因预测** | `multiMiR` (R 包) / `TargetScan` | miRNA 本身无功能，必须结合 3'UTR 种子序列预测其调控的靶基因。推荐多数据库取交集以降低假阳性。 |
