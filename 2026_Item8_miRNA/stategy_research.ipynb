目前（2026年4月）miRNA测序（小RNA-seq）的最主流方法是：
Illumina短读长测序平台 + 商用gel-free/UMI高通量文库制备试剂盒（以 QIAseq miRNA Library Kit 和 Illumina miRNA Prep 为代表）。
这是全球绝大多数核心实验室、测序公司和服务商的标准方案，占据90%以上的实际应用份额。相比传统方法，它大幅降低了操作难度、减少了PCR偏好和adapter dimer污染，适合低输入量（1 ng总RNA）和高通量（384–768样本/次）。
下面从样品准备 → 文库制备 → 测序 → 数据分析全流程详细说明（重点在测序仪器和数据分析策略）。

1. 样品准备（Sample Preparation）
输入材料：

总RNA：1–500 ng（最低可达0.1–1 ng，视试剂盒而定）。
推荐样本类型：组织、细胞、血浆/血清、外泌体、FFPE（福尔马林固定石蜡包埋）等。
质量要求：RNA Integrity Number (RIN) ≥ 7（组织/细胞）；生物液体样本可略低，但需去除rRNA/tRNA污染。

关键步骤：

RNA提取：使用TRIzol、miRNeasy Mini Kit（Qiagen）或柱式试剂盒。
小RNA富集（可选）：对于总RNA直接上机的主流方案，已可省略传统胶回收。
质量控制：Qubit + Bioanalyzer（检查小RNA峰 ~15–30 nt）或QIAseq miRNA QC Kit。

主流趋势：2026年主流是直接用总RNA低输入量，不再强制胶大小选择（gel excision），以减少样品损失和偏差。
2. 文库制备（Library Preparation）——最关键环节
最主流试剂盒（2026年排名前三）：

QIAseq miRNA Library Kit（Qiagen）：当前使用最广泛，支持UMI（Unique Molecular Index），可有效去除PCR重复，适合低输入和生物液体样本。
Illumina miRNA Prep：官方gel-free工作流，操作最简单，3小时上手时间，支持384个独特双索引。
NEBNext Small RNA Library Prep Set / Takara SMARTer smRNA-Seq Kit：备选，偏好低输入或特定样本。

文库制备核心流程（以QIAseq/Illumina miRNA Prep为例）：

3'接头连接（3' Adapter Ligation）：使用预腺苷酸化接头，高效捕获小RNA。
5'接头连接（5' Adapter Ligation）：部分试剂盒（如QIAseq）先逆转录再接头，减少dimer。
逆转录（RT） + PCR扩增（带索引）。
纯化：柱式纯化或磁珠纯化（无需胶切）。
文库质控：Bioanalyzer/TapeStation检查峰值 ~140–160 bp，浓度 > 2 nM。

优势：高通量（768索引）、低偏好、UMI去重复、gel-free（节省时间、减少损失）。
3. 测序仪器（Sequencing Platforms）——重点
最主流平台（2026年）：

Illumina NovaSeq X / NovaSeq X Plus：绝对主流，高通量（一次可测数百个样本）。
NextSeq 2000 / NextSeq 1000：中通量实验室首选。
MiSeq / iSeq 100：小项目或验证用。

推荐测序参数：

读长：1×50 bp 或 1×75 bp 单端测序（Single-end, SE）。小RNA只需读取前50–75 bp即可覆盖整个miRNA。
深度：每个样本 5–10 百万 reads（M reads）即可满足定量需求（差异表达分析）；发现新miRNA可提高到15–20 M reads。
运行模式：NovaSeq X的P2/P3/P4 flow cell（XLEAP-SBS化学），产出极高，成本低。

为什么是Illumina？

SBS（Sequencing by Synthesis）技术仍占全球测序数据90%以上。
小RNA无需长读长（PacBio / Nanopore目前在miRNA上仍为小众/补充）。

4. 数据分析策略（Data Analysis）——重点
主流分析流程（2026年标准管道）：

原始数据质控与预处理（~1–2小时）
FastQC / MultiQC → 质量评估。
fastp / Cutadapt：去除3'/5'接头、polyA、低质量读段。
UMI去重（如果用了QIAseq等带UMI的试剂盒）。

比对与定量（核心步骤）
主流工具：
miRDeep2（最经典，兼顾已知miRNA + 新miRNA发现）。
ShortStack（植物miRNA优秀）。
miND pipeline（2022–2025优化版，集成AI）。
ENCODE small RNA pipeline（官方标准化）。

参考数据库：miRBase v22/v23（成熟体 + 前体）。
比对器：Bowtie / BWA-aln（允许1–2个错配）。

差异表达与下游分析（重点）
定量：使用miRNA counts矩阵。
差异表达：DESeq2 或 edgeR（金标准，支持UMI校正）。
新miRNA发现：miRDeep2 score过滤。
功能分析：
miRTarBase / TargetScan（靶基因预测）。
DIANA-miRPath / miRPathDB（通路富集）。
clusterProfiler（GO/KEGG）。


高级/可选分析
isomiR分析（miRNA变体）。
外泌体/循环miRNA特异管道。
单细胞miRNA（新兴，但仍非主流）。


推荐完整管道（一键式）：

nf-core/smrnaseq（Nextflow管道，最推荐）。
miND 或 CAP-miRSeq。
Galaxy / BaseSpace（Illumina云平台）一键分析。

总结：2026年最主流完整方案

样品 → 总RNA（1–500 ng）
文库 → QIAseq miRNA 或 Illumina miRNA Prep（gel-free + UMI）
测序 → Illumina NovaSeq X / NextSeq 2000，1×50–75 bp，5–10 M reads/样本
分析 → fastp → miRDeep2 / ShortStack → DESeq2 → miRTarBase/miRPath
