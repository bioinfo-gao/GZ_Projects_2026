| 面板类型 (Panel Type) | 典型 Plex 数量 | 核心设计逻辑 | 探针优化重点 (Probe Optimization) | 典型应用场景 |
| :--- | :--- | :--- | :--- | :--- |
| **RUO Products** | 300 - 1000+ | 广泛覆盖主要细胞群与核心通路 | 平衡高低表达基因，防止光学拥挤 | 商业化标准品，健康/疾病组织 Atlas 测序 |
| **Custom Designs** | 50 - 300 | 基于客户自测 scRNA-seq Marker | 快速特异性脱靶筛查，可灵活增删靶点 | 实验室独家靶标验证，特殊罕见病研究 |
| **FFPE Optimization** | 100 - 500 | 避开易降解靶点，选取高丰度特征基因 | **极短靶向区域 (<70bp)**，增加探针冗余度 | 临床回顾性研究，医院库存切片分析 |
| **3D Assays** | 50 - 200 | 强化结构性、形态学及边界 Marker | 高光稳定性染料，深层组织渗透性设计 | 大脑核团 3D 建模，全胚胎发育轨迹追踪 |
| **Disease-Specific** | 200 - 500 | 模块化通路设计 (如免疫耗竭、微环境) | 覆盖异构体，可能包含融合基因/SNP 探针 | 靶向药效力评估，肿瘤微环境空间解析 |

| 评分维度 | 核心量化指标 (Metrics) | 最佳通过阈值 (Pass Criteria) | 权重 / 淘汰机制 |
| :--- | :--- | :--- | :--- |
| **1. Target Ranking** | Log2FC, FDR, Pct_diff, 归一化表达量 | Log2FC > 1.0, FDR < 0.01 | 高权重 (决定 Panel 的生物学价值) |
| **2. Sequence Complexity**| GC%, $T_m$, $\Delta G$, 连续重复碱基 | GC: 40-60%, 无 >4 连续同聚物 | **一票否决** (理化不达标直接废弃) |
| **3. Barcode Compatibility**| Hamming 距离, 转录组同源性 (BLAST) | Hamming 距离 $\ge 3$, 无同源 | 高权重 (决定能否成功解码) |
| **4. Tissue Performance** | 共有外显子占比, 序列连续性 (FFPE) | 靶向区 < 60bp, 避开剪接位点 | 中权重 (视样本类型动态调整) |
