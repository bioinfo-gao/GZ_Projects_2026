一、完整 Panel 自动化设计 Pipeline

适用于：

Amplicon NGS

Hybrid Capture

RUO 或 IVD 早期版本

🔷 总体架构
Input: Gene List / Variant List
    ↓
Biological Ranking Engine
    ↓
Region Extraction Engine
    ↓
Sequence Quality Filtering
    ↓
Primer Design Engine
    ↓
Multiplex Optimization Engine
    ↓
FFPE Adaptation (optional)
    ↓
In-silico Specificity Simulation
    ↓
Scoring & Ranking
    ↓
Export: BED / FASTA / Excel / JSON
1️⃣ Step 1 – 输入层

支持：

Gene symbol list

Transcript ID

COSMIC hotspot list

BED file

客户自定义变异

数据来源接口：

Ensembl GTF

RefSeq

COSMIC

ClinVar

2️⃣ Step 2 – Target Ranking Engine

输入字段：

| Gene | Mutation freq | Expr | Clinical | Pathway | |

输出：

FinalTargetScore = 0.4 Mutation +
                   0.2 Clinical +
                   0.2 Biology +
                   0.2 Expression

阈值过滤：

if score < 0.6 → drop
3️⃣ Step 3 – Region Extraction
Amplicon 模式

规则：

覆盖 exon ±20 bp

Hotspot ±50 bp

CNV gene → 全 CDS

输出：BED 文件

4️⃣ Step 4 – 序列复杂度过滤

过滤条件：

GC < 35% → penalty
GC > 70% → penalty
Homopolymer > 5 → reject
RepeatMasker overlap > 40% → reject
5️⃣ Step 5 – Primer 设计引擎

使用：

Primer3 API

自定义 Python wrapper

参数模板：

PRIMER_OPT_SIZE=20
PRIMER_MIN_SIZE=18
PRIMER_MAX_SIZE=24
PRIMER_OPT_TM=60
PRIMER_MIN_TM=58
PRIMER_MAX_TM=62
PRIMER_PRODUCT_SIZE_RANGE=80-150
6️⃣ Step 6 – Multiplex 优化

建立 compatibility matrix：

计算：

Primer-dimer ΔG

Cross complementarity

Tm variance

GC variance

构建图模型：

节点 = amplicon

边 = 不兼容

图染色算法 → 分成多个 PCR pool

7️⃣ Step 7 – In silico Specificity

BLAST against hg38

允许 ≤1 mismatch

非目标结合 >1 → reject

8️⃣ Step 8 – 输出结构

生成：

二、FFPE 优化模板（工业 SOP 模板）

FFPE 是独立工程问题。

🔷 FFPE 特性

DNA fragmented (~100–300 bp)

C>T deamination

Crosslinking

Low input

🔷 FFPE Panel 设计规则
1️⃣ Amplicon 长度
Optimal: 70–110 bp
Max: 120 bp
2️⃣ Primer 设计约束
Avoid CpG at 3' end
Avoid C at 3' terminal
GC clamp ≤2
No >4 G run
3️⃣ 质量控制模块

必须加入：

QC 类型	目的
Short fragment control (80 bp)	DNA integrity
Medium fragment (150 bp)	fragmentation ratio
Housekeeping multi-exon	RNA quality
4️⃣ FFPE 风险评分
FFPE_Risk =
0.4 * Amplicon_Length_Penalty +
0.3 * GC_Extreme +
0.2 * CpG_Density +
0.1 * Structure
5️⃣ UDG 适配

如果做 IVD：

建议加入 UDG 预处理模块

Variant caller 加入 C>T artifact filter

Full exon (top genes)

3000–6000 amplicons

3–4 pools

大 Panel (~500 gene)

Mixed design

8000–12000 amplicons

4–6 pools

建议 hybrid capture

🔷 设计容量估算

经验公式：

Amplicon_count ≈ (Total_target_bp / 120)

例如：

500 genes × avg 1500 bp CDS
= 750,000 bp

750,000 / 120 ≈ 6250 amplicons

🔷 Pool 划分逻辑

约束：

每 pool ≤1500 amplicons

Tm variance ≤1.5°C

GC variance ≤10%

No strong cross-dimer
