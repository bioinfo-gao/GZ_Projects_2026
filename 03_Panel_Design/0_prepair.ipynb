```python
from Bio.Seq import Seq
# from Bio.SeqUtils import GC
from Bio.SeqUtils import MeltingTemp as mt

def gc_content(primer):
    gc_count = primer.count('G') + primer.count('C')
    return (gc_count / len(primer)) * 100

def design_primers(sequence, primer_length=20, gc_range=(40, 60), tm_range=(58, 62)):
    candidates = []

    if len(sequence) < primer_length:
        raise ValueError("Sequence length must be greater than or equal to primer length.")

    for i in range(len(sequence) - primer_length + 1):
        primer = sequence[i:i+primer_length]
        gc = gc_content(primer)
        tm = mt.Tm_Wallace(primer)

        if gc_range[0] <= gc <= gc_range[1] and tm_range[0] <= tm <= tm_range[1]:
            candidates.append({
                "sequence": primer,
                "start": i,
                "length": primer_length,
                "GC%": round(gc, 2),
                "Tm": round(tm, 2)
            })

    return candidates

# 示例用法
sequence = "ATGCGTACGTAGCTAGCTAGCTGACTGATCGATCGTACGTAGCTAGCTAGCTAGC"
primers = design_primers(sequence)

print(f"Found {len(primers)} primer candidates:")

for p in primers[:5]:
    print(p)
```

```python
#进阶版本：避免二聚体（简单检测）
def has_self_complementarity(seq):
    rev_comp = str(Seq(seq).reverse_complement())
    return seq[-4:] in rev_comp

filtered = []

for p in primers:
    if not has_self_complementarity(p["sequence"]):
        filtered.append(p)

for item in filtered:
    print(item)
print("Filtered primers:", len(filtered))
```

评分公式
归一化到 0–1：
Score=∑wi⋅xi  
$Score = \sum w_i \cdot x_i$

示例权重（可调）： 

Biology: 0.25  
Clinical: 0.20  
Expression: 0.15  
Mutation: 0.20  
Network: 0.10  
Resistance: 0.10

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

weights = {
    "biology": 0.25,
    "clinical": 0.20,
    "expression": 0.15,
    "mutation": 0.20,
    "network": 0.10,
    "resistance": 0.10
}

def rank_targets(df):
    scaler = MinMaxScaler()
    scaled = scaler.fit_transform(df[list(weights.keys())])
    df_scaled = pd.DataFrame(scaled, columns=weights.keys())

    df["final_score"] = sum(
        df_scaled[col] * weight for col, weight in weights.items()
    )

    return df.sort_values("final_score", ascending=False)
```

Shannon Entropy 公式

$H = -\sum p_i \log_2(p_i)$

```python
import math
from collections import Counter

def shannon_entropy(seq):
    counts = Counter(seq)
    total = len(seq)
    entropy = 0

    for base in counts:
        p = counts[base] / total
        entropy -= p * math.log2(p)

    return entropy

def sequence_complexity_score(seq):
    gc = (seq.count("G") + seq.count("C")) / len(seq)
    entropy = shannon_entropy(seq)

    gc_score = 1 - abs(gc - 0.5)
    entropy_score = entropy / 2  # max approx 2

    return 0.6 * entropy_score + 0.4 * gc_score
```
