在 WSL 远程设置中“死磕”路径
Positron 目前最稳定的手动路径注册方式是使用数组格式。
按下 Ctrl + Shift + P。
输入并选择：Preferences: Open Remote Settings (JSON) (WSL: Ubuntu)。
确保文件内容精准如下（注意：使用 customBinaries 这个 Key）：

{   
    // JSON 的注释采用C 格式， // or /* 内容 */
    "positron.r.customBinaries": [
        "/home/zhen/miniforge3/envs/seurat_env/bin/R", // single cell analysis R 4.3.3
        "/home/zhen/miniforge3/envs/spatial_R/bin/R"   // spatial analysis R 4.5.1
    ],
    "positron.r.customRootFolders": [
        "/home/zhen/miniforge3/envs/seurat_env/bin",
        "/home/zhen/miniforge3/envs/spatial_R/bin"   // no comma in the last item
    ]                                                // no comma in the last section 
}

可能需要重启，新的R环境可能才能识别 是另外一种强制识别的办法

第三步：强制重新加载窗口
配置保存后，必须强制 Positron 重新扫描：
按下 Ctrl + Shift + P。
输入并选择：Developer: Reload Window。

验证与手动选择
重新加载后，请点击 Positron 右上角的解释器选择区域（原本显示 Python 的地方）：
在弹出的列表中，观察是否出现了 spatial_R 或 R 4.5.1。
如果出现了：点击选择它，右侧的 Console 应该会自动启动 R Session。
如果还是只有 Python：
点击列表最下方的 + Add Runtime...。
在弹出的文本框中，手动粘贴路径：/home/zhen/miniforge3/envs/spatial_R/bin/R。
